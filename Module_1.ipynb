# Module 1 – Distributed Computing: An Introduction  

Welcome to our first module on **Distributed Computing**.  
In this module, we’ll explore what distributed computing means, where it came from, and why it’s the backbone of modern systems such as the Internet, cloud computing, and mobile apps.  

---

## 1.1 Key Definitions  

Before diving deep, we need a **shared vocabulary**.  
Distributed computing has evolved rapidly, and many terms are used differently in different contexts.  

### 1. Early (Monolithic) Computing  
- Early computers used **a single CPU** to execute all tasks.  
- Called **uni-processor** or **monolithic computing**.  
- Example: 1960s **mainframes** served entire organizations.  
- **Limitations**: Expensive, slow to scale, single point of failure.  

### 2. Distributed System  
- A **collection of independent computers** connected by a network.  
- They collaborate to perform tasks **as if they were one system**.  
- Independent → no shared memory or execution space.  
- Two categories:  
  - **Loosely coupled**: Connected only via network (e.g., cloud servers).  
  - **Tightly coupled**: Share memory/resources directly (e.g., supercomputers).  
- **Example**: Google search engine – feels like one service, but runs on thousands of machines.  

### 3. Request for Comments (RFCs)  
- **Public documents** for Internet standards.  
- Define protocols like **HTTP (web)**, **SMTP (email)**, **FTP (file transfer)**.  
- Why important? Distributed systems **must agree on rules** to communicate.  

### 4. ARPANET (1970)  
- First operational packet-switched network.  
- Predecessor of today’s Internet.  
- Initially connected research labs and universities.  

### 5. Distributed Computing  
- Simply: **computing performed in a distributed system**.  
- Programs on different computers **work together** via the network.  
- Example: Online banking →  
  - Web server (interface)  
  - Database server (transactions)  
  - Authentication server (security)  

### 6. Client-Server Model  
- **Server**: Provides services (e.g., website, database, email).  
- **Client**: Requests and uses those services.  
- Foundation of most distributed systems, including cloud services.  

### 7. Services vs. Applications  
- **Network Services**: Fundamental offerings like Web, Email, FTP.  
- **Network Applications**: User-facing apps like shopping carts, auctions, games.  
- Distinction is fuzzy → the **Web** is both a service and a platform for applications.

## 1.2 A Brief History of Distributed Computing  

### 1960s – Standalone Systems  
- Each computer worked **independently**.

- **Email (1972)**: First true distributed application on ARPANET.  
- **File Transfer Protocol (FTP)**: Automated file sharing.  
- Idea: computers could **collaborate** instead of working alone.  

### 1980s – Growth of Network Services  
- Email and FTP became widely used.  
- Universities and governments embraced networking.  
- Research into **distributed file systems** and **remote procedure calls**.  

### 1990s – The World Wide Web  
- Invented by **Tim Berners-Lee at CERN**.  
- Introduced **hypertext (links)** → made information sharing intuitive.  
- Explosive growth in Internet hosts and domains.  
- Enabled **search engines, e-commerce, online communities**.

### 1. Monolithic Computing  
- **Definition:** Computing performed on a single machine with no network dependency.  
- **Single-user monolithic computing** → One user, one system (e.g., running MS Word on a standalone PC).  
- **Multi-user monolithic computing (Timesharing):**  
  - Popular in **1970s–1980s** (e.g., IBM 360, Univac 1100).  
  - One large **mainframe** shared by many users via **terminals**.  
  - Used for payroll, billing, and university administration.  

**Telecom context:** Early **telephone exchanges and switching systems** relied on centralized mainframes.  
**Information context:** Data records and transaction systems (banking, payroll) were centralized, limiting scalability.  

---

### 2. Distributed Computing  
- **Definition:** Multiple **independent networked computers** collaborate on tasks.  
- Each computer has its **own CPU & resources**.  
- Users access both **local resources** and **remote resources** through the network.  
- **Example:** The **World Wide Web** – browsers (clients) interact with **remote servers** to fetch web pages.  

**Telecom context:** Distributed base stations in **cellular networks (4G/5G)**, where processing is shared across nodes.  
**Information context:** **Cloud services** like Gmail, Dropbox – run on distributed servers ensuring scalability and availability.

- **Definition:** Multiple processors execute a single program simultaneously.  
- Typically within **one machine with multiple CPUs**, but can extend to clusters using special distributed processing software.  
- **Advantages:** High speed, solves complex problems.  
- **Challenges:** Difficult to divide tasks without interference.

**Telecom context:** Uses **Internet as the backbone** to connect idle computers worldwide.  
**Information context:** Demonstrates how **decentralized resources** can solve massive computational problems at low cost.

## 1.4 The Strengths and Weaknesses of Distributed Computing  

---

### Why This Matters  
As telecom and information engineers, you’ll be working on **large-scale systems** like:  
- **5G/6G networks** that must scale to millions of users.  
- **Cloud platforms** that need reliability, even if servers fail.  
- **Cybersecurity for telecom infrastructure** that is exposed to attacks.  

Understanding the **trade-offs of distributed computing** helps you design robust, scalable, and secure systems.  

---

## Strengths of Distributed Computing  

### 1. Affordability & Ubiquity  
- Personal computers and servers today are **cheaper and more powerful** than mainframes of the 1970s.  
- Internet access is now **affordable and widespread**, enabling massive distributed systems.  

**Telecom context:**  
- Deployment of **base stations and edge devices** is feasible because hardware costs are low.  
- Telecom operators can distribute processing across the network affordably.  

---

### 2. Resource Sharing  
- Organizations maintain local systems but **share resources** over networks.  
- Example: The Web allows document and resource sharing across companies.  

**Information Engineering context:** Shared **databases, APIs, and services** (e.g., cloud storage, distributed ledgers).  
**Telecom context:** Shared **network infrastructure** among operators (e.g., tower sharing).  

---

### 3. Scalability  
- Monolithic computing is limited to one system.  
- Distributed computing scales **horizontally**: add more machines → handle more workload.  
- Example: Email services add more mail servers to handle rising demand.  

**Telecom context:** Scaling **user connections** in 5G/6G by adding more distributed radio units (RUs).  
**Information context:** Scaling **web traffic** with content delivery networks (CDNs).  

---

### 4. Fault Tolerance  
- Resources can be **replicated/mirrored** across nodes.  
- If one node fails, others continue service (e.g., redundant databases).  
- Key research area: achieving **maximum reliability despite failures**.  

**Telecom context:** Call continuity in **mobile handovers** when one base station fails.  
**Information context:** Data replication in **distributed databases** (e.g., Google Spanner, Cassandra).  

---

## Weaknesses of Distributed Computing  

### 1. Multiple Points of Failure  
- More nodes = more chances of failure.  
- Failure of one unknown node can still break the system.  
- Leslie Lamport:  
  *“A distributed system is one in which the failure of a computer you didn’t even know existed can render your own computer unusable.”*  

**Telecom context:** One faulty **fiber link or router** can disrupt services for thousands of users.  
**Information context:** A crashed microservice in cloud architecture can halt dependent applications.  

---

### 2. Security Concerns  
- Distributed systems are **decentralized**, often spanning multiple organizations.  
- Harder to enforce **uniform security policies**.  
- Vulnerable to attacks: worms, viruses, DDoS (Denial of Service).  

**Examples:**  
- Yahoo, eBay, CNN hit by DDoS attacks (2000) → networks flooded with fake traffic.  
- Attack types:  
  - **SYN flood** – attackers overload servers with fake connection requests.  
  - **Smurf attack** – attackers use third-party machines to amplify traffic.  

**Telecom context:** Telecom networks are prime targets for DDoS attacks, especially against **5G cores** and **DNS services**.  
**Information context:** Cloud applications are exposed to **SQL injection, ransomware, and API attacks**.  

---

## Mitigation Strategies  

- **Fault tolerance:** Redundancy, load balancing, failover clusters.  
- **Security techniques:**  
  - Encryption, authentication, authorization.  
  - Digital certificates & signatures.  
  - Firewalls, intrusion detection, sandboxes.  
- **Collaborative defense:** ISPs, telecom providers, and enterprises must coordinate to block attacks.  

---

## Key Takeaways  
- Distributed computing is **affordable, scalable, fault-tolerant, and resource-efficient**.  
- But it introduces **complexity, security challenges, and failure risks**.  
- For telecom and information engineers, mastering these trade-offs is essential for:  
  - **Designing reliable 5G/6G networks.**  
  - **Building secure and scalable cloud-based applications.**

## 4. Concurrency in Distributed Systems  

Concurrency allows multiple tasks to be executed at the same time.  

---

### 4.1 Multiple Computers  

Processes running on different machines, communicating via the network.  

 Example: A 5G base station talking to a core network server.

### 4.2 Multiple Processes on One Machine  

The OS supports multitasking:  

- True parallelism on multi-core CPUs  
- Time-slicing on single-core CPUs  


````python
import multiprocessing

def worker(name):
    print(f"Process {name} is running")

if __name__ == "__main__":
    p1 = multiprocessing.Process(target=worker, args=("A",))
    p2 = multiprocessing.Process(target=worker, args=("B",))
    p1.start()
    p2.start()
    p1.join()
    p2.join()
````

# Two Processes Run Concurrently

## 4.2 Multiple Threads in One Process

- A single process can contain **multiple threads**.
- Threads share the same **memory space** but execute **independently**.
- Benefits of multithreading:
  - Efficient **CPU utilization**
  - Faster **task execution**
  - Simplified **program structure** for concurrent tasks
- Key considerations:
  - **Synchronization** to avoid race conditions
  - Proper **thread management** to prevent deadlocks


````python
import threading

def task(name):
    for i in range(3):
        print(f"Thread {name} is working...")

if __name__ == "__main__":
    t1 = threading.Thread(target=task, args=("X",))
    t2 = threading.Thread(target=task, args=("Y",))
    t1.start()
    t2.start()
    t1.join()
    t2.join()
````

- Threads **share memory**, which is useful in telecom applications for:
  - Handling **multiple calls**
  - Managing **messages simultaneously**

# 5. Parent and Child Processes

- A **parent process** can spawn **child processes** to handle **subtasks**.


````python
import multiprocessing

def child_task():
    print("Child process is working...")

if __name__ == "__main__":
    parent = multiprocessing.current_process()
    print("Parent process:", parent.name)
    child = multiprocessing.Process(target=child_task)
    child.start()
    child.join()
````

- A **call server (parent)** may spawn **child processes** to handle:
  - **Authentication**
  - **Billing**
  - **Signaling**

# 6. Race Conditions and Synchronization

- A **race condition** occurs when multiple threads or processes **modify shared data at the same time**.

### Example (Race Condition)


````python
import threading

counter = 0

def increment():
    global counter
    for _ in range(100000):
        counter += 1

t1 = threading.Thread(target=increment)
t2 = threading.Thread(target=increment)

t1.start()
t2.start()
t1.join()
t2.join()

print("Final counter value:", counter)  # ❌ Often less than expected
````

# Solution to Race Conditions

- **Use Locks** to synchronize access to shared resources


````python
import threading

counter = 0
lock = threading.Lock()

def increment():
    global counter
    for _ in range(100000):
        with lock:   # Ensures only one thread modifies counter at a time
            counter += 1

t1 = threading.Thread(target=increment)
t2 = threading.Thread(target=increment)

t1.start()
t2.start()
t1.join()
t2.join()

print("Final counter value:", counter)  #  Correct result
````

# Preventing Billing Errors

- **Scenario:** Two processes updating a subscriber’s balance simultaneously.
- **Solution:** Use **synchronization mechanisms** (e.g., locks) to prevent errors.

# 1. Protocols

- A **protocol** is like the **grammar of digital communication**.
- Defines rules for:
  - **Starting** a conversation
  - **Transmitting** data
  - **Ending** communication
- Without protocols, devices **cannot understand each other**.

### Contexts
- **Telecom:** 4G/5G call setup uses **SIP protocol**; phones must follow signaling rules.
- **Information Engineering:** Browsers use **HTTP/HTTPS** to request pages.

### Python Demo
- Simulated **protocol** using **turn-taking**


````python
import time

def simple_protocol(sender, receiver, messages):
    print("=== Simulated Protocol Conversation ===\n")
    for msg in messages:
        print(f"{sender} → {receiver}: {msg}")
        time.sleep(0.5)
        print(f"{receiver} → {sender}: ACK (received)\n")
        time.sleep(0.5)

simple_protocol("Client", "Server", ["Hello", "Request: Data", "Goodbye"])
````

- **Request–Response Rules:** Maintain order in communication, similar to how protocols work.

# 2. Network Architectures

- **Purpose:** Manage complexity by dividing networking into **layers**.

## Models
- **OSI Model (7 layers):** Theoretical, mainly used in textbooks.
- **Internet Model (4 layers):** Practical, used in real systems.

| Layer        | Function                    | Example         |
|--------------|----------------------------|----------------|
| Application  | Services                   | HTTP, FTP, SMTP|
| Transport    | Process-to-process delivery | TCP, UDP       |
| Internet     | Addressing & routing        | IP             |
| Physical     | Transmission medium         | Fiber, Wi-Fi   |

### Contexts
- **Telecom:** 5G base stations use **layered protocols** for seamless video calls.
- **Information Engineering:** Distributed databases replicate data using **layered communication**.

### Python Demo
- Demonstrates **Encapsulation & Decapsulation**


````python
layers = ["Application", "Transport", "Internet", "Physical"]

def send_message(message):
    print(f"Original Message: {message}\n")

    # Encapsulation (sender side)
    for layer in reversed(layers):
        message = f"[{layer}] {message}"
        print(f"↓ {layer} adds header: {message}")

    print("\n--- Transmission ---\n")

    # Decapsulation (receiver side)
    for layer in layers:
        message = message.replace(f"[{layer}] ", "", 1)
        print(f"↑ {layer} removes header: {message}")

    print(f"\nFinal Message Delivered: {message}")

send_message("Hello Distributed Systems!")
````

- Observe how data is **wrapped with headers** and **unwrapped layer by layer**.

# 3. Transport Protocols: TCP vs UDP

- **TCP (Transmission Control Protocol)**
  - Reliable: guarantees delivery
  - Retransmits lost data

- **UDP (User Datagram Protocol)**
  - Fast but unreliable
  - Commonly used for **live video** or **VoIP**

### Contexts
- **Telecom:** VoIP often uses UDP – better to lose a packet than delay speech.
- **Information Engineering:** File transfers (FTP, HTTP) use TCP – every byte must arrive intact.

### Python Demo
- Simulates **TCP vs UDP behavior**


````python
import random

def udp_send(messages):
    print("UDP Transmission (unreliable):")
    for msg in messages:
        if random.random() > 0.3:  # 70% chance delivered
            print(f"Delivered: {msg}")
        else:
            print(f"Lost: {msg}")
    print()

def tcp_send(messages):
    print("TCP Transmission (reliable):")
    for msg in messages:
        delivered = False
        while not delivered:
            if random.random() > 0.3:
                print(f"Delivered: {msg}")
                delivered = True
            else:
                print(f"Retrying {msg}...")

msgs = ["Packet1", "Packet2", "Packet3"]
udp_send(msgs)
tcp_send(msgs)
````

- This is to allow you see the **trade-off between reliability and speed**.

# 1.7 Network Architecture Protocols

- In **distributed computing**, understanding **network protocols** and **communication methods** is crucial.
- Computers communicate using **protocols**, which define how data is **sent, received, and interpreted**.

---

## Internet Protocol Stack

The Internet architecture is divided into **layers**, each with specific protocols:

- **Internet Layer:** IP (Internet Protocol)  
  - Responsible for **addressing** and **routing packets** to remote computers.

- **Transport Layer:**  
  - **TCP (Transmission Control Protocol):** Connection-oriented, ensures **ordered and reliable delivery**.  
  - **UDP (User Datagram Protocol):** Connectionless, **faster**, may **lose packets**.

- **Application Layer:**  
  - Protocols like **HTTP, FTP, SNMP** support **end-user applications**.

- These layers collectively form the **TCP/IP stack**, which underlies almost all Internet communication.

---

## Simulating the TCP/IP Stack in Python

- Visualizes how a **message is encapsulated** at each layer of the network and then **decapsulated** at the receiver.


````python
layers = ["Application (HTTP/FTP)", "Transport (TCP/UDP)", "Internet (IP)", "Physical"]

def send_message(message, protocol="TCP"):
    print(f"\nOriginal Message: {message}\n")

    # Encapsulation
    for layer in reversed(layers):
        header = f"[{layer}]"
        if layer.startswith("Transport"):
            header += f"-{protocol}"
        message = f"{header} {message}"
        print(f"Encapsulated: {message}")

    print("\n--- Transmission ---\n")

    # Decapsulation
    for layer in layers:
        message = message.replace(f"[{layer}] ", "", 1)
        if layer.startswith("Transport"):
            message = message.replace(f"-{protocol} ", "", 1)
        print(f"Decapsulated at {layer}: {message}")

    print(f"\nDelivered Message: {message}\n")

send_message("Hello Distributed Systems!", protocol="TCP")
send_message("Quick Update Packet", protocol="UDP")
````

# Insight

- **TCP:** Adds headers to ensure reliability; packets are **tracked and reordered** if necessary.
- **UDP:** Adds minimal headers; packets may **arrive out of order** or be **lost**.

---

# Connection-Oriented vs Connectionless Communication

## Connection-Oriented Communication (TCP-like)
- A **dedicated connection** is established before data transfer.
- Ensures **packets arrive in order** and **without loss**.
- Overhead: **connection setup and teardown**.
- Examples: phone calls, file transfers over TCP.

## Connectionless Communication (UDP-like)
- **No prior connection** is required; each packet is independent.
- Faster but may result in **lost or out-of-order packets**.
- Examples: email, video streaming, DNS requests.

### Python Simulation
- Demonstrates **Connection-Oriented vs Connectionless Communication**


````python
import random
import time

def connection_oriented(messages):
    print("=== Connection-Oriented (TCP-like) ===")
    print("Establishing connection...")
    time.sleep(1)
    for msg in messages:
        print(f"Sending: {msg}")
        time.sleep(0.5)
        print(f"ACK received for: {msg}")
    print("Connection closed.\n")

def connectionless(messages):
    print("=== Connectionless (UDP-like) ===")
    for msg in messages:
        if random.random() > 0.2:  # 80% chance delivered
            print(f"Delivered: {msg}")
        else:
            print(f"Lost: {msg}")
        time.sleep(0.5)
    print("Session ended.\n")

messages = ["Packet1", "Packet2", "Packet3", "Packet4"]
connection_oriented(messages)
connectionless(messages)
````

# Insight

- **TCP:** Guarantees delivery but incurs **overhead**.
- **UDP:** Faster, suitable for **small or occasional messages**, but **reliability is not guaranteed**.

# Dynamic Visualization of TCP vs UDP Delivery

- Packet delivery can be **visualized dynamically** using **matplotlib animations**.


````python
import matplotlib.pyplot as plt
import matplotlib.animation as animation

fig, ax = plt.subplots()
ax.set_xlim(0, 10)
ax.set_ylim(0, 2)
ax.set_yticks([0.5, 1.5])
ax.set_yticklabels(['UDP', 'TCP'])
line_tcp, = ax.plot([], [], 'go', label='TCP')
line_udp, = ax.plot([], [], 'ro', label='UDP')
ax.legend()

tcp_packets = [1, 2, 3, 4, 5]
udp_packets = [1, 2, 3, 4, 5]

def animate(frame):
    line_tcp.set_data(tcp_packets[:frame+1], [1.5]*(frame+1))
    line_udp.set_data([p for p in udp_packets[:frame+1] if random.random() > 0.2], [0.5]*len(udp_packets[:frame+1]))
    return line_tcp, line_udp

ani = animation.FuncAnimation(fig, animate, frames=len(tcp_packets), interval=800, blit=True)
plt.show()
````

# Insight

- **TCP packets (green):** Always arrive — demonstrates **reliability**.
- **UDP packets (red):** May drop — illustrates **connectionless behavior**.

---

# Key Take away

- **Protocols** define communication rules and are **distinct from implementations**.
- The **TCP/IP stack** has multiple layers; each layer adds **headers and functionality**.
- **TCP vs UDP** demonstrates the **trade-off between reliability and speed**.
- **Connection-oriented vs connectionless communication** is fundamental for understanding **distributed applications**.

# Connection-Oriented vs Connectionless Communication

## Connection-Oriented (TCP-like)
- Suitable for **large data streams** or **multiple rounds of communication**
- Guarantees **delivery and order**

## Connectionless (UDP-like)
- Suitable for **small messages** with limited rounds
- Faster but may **lose packets** or deliver them **out of order**

### Python Simulation
- Demonstrates **connection-oriented vs connectionless communication**


````python
def simulate_network(messages, mode="TCP"):
    print(f"\nSimulating {mode}-like communication")
    for msg in messages:
        if mode == "TCP":
            print(f"Sending: {msg} -> ACK received")
        else:  # UDP
            if random.random() > 0.2:
                print(f"Delivered: {msg}")
            else:
                print(f"Lost: {msg}")
messages = ["Data1", "Data2", "Data3"]
simulate_network(messages, "TCP")
simulate_network(messages, "UDP")
````

# Host Identification and IPv4 Addresses

- Every Internet **host** must have a **unique address**.
- **IPv4 addresses:** 32-bit numbers (over 4 billion addresses)
- Addresses are divided into **classes (A–E)** to manage networks of different sizes:

| Class | Prefix | Networks     | Hosts per Network |
|-------|--------|-------------|-----------------|
| A     | 0      | 128         | 16 million      |
| B     | 10     | 16,384      | 65,536          |
| C     | 110    | ~2 million  | 256             |
| D     | 1110   | Multicast   | N/A             |
| E     | 1111   | Reserved    | N/A             |




````python
ip_binary = "10000001010000010001100000110010"  # 129.65.24.50
ip_decimal = ".".join(str(int(ip_binary[i:i+8], 2)) for i in range(0, 32, 8))
print(f"Dotted Decimal IP: {ip_decimal}")
````

# 1.9 Identifying Processes with Protocol Ports and Name Resolution

- Locating a host using its **IP address** or **domain name** is only part of network communication.
- To deliver data to a **specific process**, we need a system for **identifying processes**.

---

## Protocol Ports

- The **transport layer (TCP/UDP)** delivers data to the **correct process**.
- Each host has **65,536 ports** (0–65,535) for TCP and UDP.
- **Well-known services:** Ports 0–1023  
  - Examples: HTTP (80), FTP (21), SMTP (25), Telnet (23)
- Applications must send data to both a **host** and a **specific port** `(host, port)`.

### Python Example
- Demonstrates **checking open ports (TCP)**


````python
import socket

def check_port(host, port):
    with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
        s.settimeout(1)
        try:
            s.connect((host, port))
            return True
        except:
            return False

host = "www.google.com"
ports = [80, 443, 21]
for port in ports:
    print(f"Port {port} open on {host}? {check_port(host, port)}")
````

## Procedural vs Object-Oriented Programming

### Procedural Programming
- Breaks programs into **procedures or functions**
- Each procedure handles a **specific task** (input, processing, output)
- Example: **C language**

### Python Example
- Demonstrates **procedural programming style**


````python
# Procedural approach
def get_numbers():
    return [1, 2, 3, 4, 5]

def compute_sum(numbers):
    return sum(numbers)

def display_result(result):
    print(f"The sum is: {result}")

numbers = get_numbers()
result = compute_sum(numbers)
display_result(result)
````

### Object-Oriented Programming (OOP)

- Uses **objects** to encapsulate **state** and **behavior**
  - **State:** Instance data (attributes)
  - **Behavior:** Methods (functions inside the object)
- Examples: **Java, Python**

### Python Example
- Demonstrates **OOP style**


````python
class Calculator:
    def __init__(self, numbers):
        self.numbers = numbers

    def compute_sum(self):
        return sum(self.numbers)

    def display_result(self):
        print(f"The sum is: {self.compute_sum()}")

calc = Calculator([1, 2, 3, 4, 5])
calc.display_result()
````

### Unified Modeling Language (UML)

- **UML:** Universal notation for specifying, visualizing, constructing, and documenting software systems.
- **Purpose:** Helps **design** and **communicate system structure**

#### Useful Diagrams
- **Class diagrams** show relationships:
  - **Inheritance:** Class B inherits from Class A
  - **Implementation:** Class B implements Interface A
  - **Dependency:** Class B depends on Class A

### Python Example
- Demonstrates **representing relationships**


````python
# Base class
class Animal:
    def speak(self):
        pass

# Inheritance
class Dog(Animal):
    def speak(self):
        return "Woof!"

# Interface (abstract class in Python)
from abc import ABC, abstractmethod

class Flyer(ABC):
    @abstractmethod
    def fly(self):
        pass

class Bird(Animal, Flyer):
    def speak(self):
        return "Chirp"

    def fly(self):
        return "Flies high"

# Usage
dog = Dog()
bird = Bird()
print(dog.speak())
print(bird.speak(), bird.fly())
````

````python
import networkx as nx
import plotly.graph_objects as go

# Define classes with attributes and methods
classes = {
    "Animal": {"type": "class", "attributes": ["age", "weight"], "methods": ["eat()", "sleep()"]},
    "Dog": {"type": "class", "inherits": "Animal", "attributes": ["breed"], "methods": ["bark()"]},
    "Bird": {"type": "class", "inherits": "Animal", "implements": "Flyer", "attributes": ["wing_span"], "methods": ["fly()"]},
    "Flyer": {"type": "interface", "methods": ["fly()"]}
}

# Create directed graph
G = nx.DiGraph()

# Add nodes and edges
for cls, props in classes.items():
    G.add_node(cls, **props)
    if "inherits" in props:
        G.add_edge(props["inherits"], cls, relation="inherits")
    if "implements" in props:
        G.add_edge(props["implements"], cls, relation="implements")

# Get layout
pos = nx.spring_layout(G, seed=42)

# Create Plotly edges
edge_x, edge_y, edge_labels = [], [], []
for u, v, data in G.edges(data=True):
    x0, y0 = pos[u]
    x1, y1 = pos[v]
    edge_x.extend([x0, x1, None])
    edge_y.extend([y0, y1, None])
    edge_labels.append(data["relation"])

edge_trace = go.Scatter(
    x=edge_x, y=edge_y,
    line=dict(width=2, color='#888'),
    hoverinfo='none',
    mode='lines'
)

# Create Plotly nodes
node_x, node_y, node_text, node_color = [], [], [], []
for node, data in G.nodes(data=True):
    x, y = pos[node]
    node_x.append(x)
    node_y.append(y)
    text = f"{node}<br>Type: {data['type']}<br>Attributes: {', '.join(data.get('attributes', []))}<br>Methods: {', '.join(data.get('methods', []))}"
    node_text.append(text)
    node_color.append('lightblue' if data['type'] == 'class' else 'lightgreen')

node_trace = go.Scatter(
    x=node_x, y=node_y,
    mode='markers+text',
    hoverinfo='text',
    textposition="bottom center",
    marker=dict(size=50, color=node_color, line=dict(width=2, color='black')),
    text=list(G.nodes),
    hovertext=node_text
)

# Build figure
fig = go.Figure(data=[edge_trace, node_trace],
                layout=go.Layout(
                    title='Interactive UML Class Diagram',
                    showlegend=False,
                    hovermode='closest',
                    margin=dict(b=20,l=5,r=5,t=40),
                    xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
                    yaxis=dict(showgrid=False, zeroline=False, showticklabels=False)
                ))

fig.show()
````

# The Architecture of Distributed Applications

- Distributed applications are often organized in a **multilayer architecture**, similar to network layers.
- A common design uses **three layers**:

---

## Presentation Layer
- Provides the **user interface**
- Example: In a shopping cart, generates **web pages viewable by a shopper**

---

## Application (Business) Logic Layer
- Handles **computation or business rules**
- Example: Computes **total order cost, tax**, and checks **credit verification**

---

## Service Layer
- Provides **underlying services** supporting the other two layers
- Examples: **Database access, directory services (DNS), interprocess communication**
- Main focus in **distributed systems**

---

## Python Interactive Example
- Visualizes a **distributed application architecture** with interactive nodes using **plotly**


````python
import networkx as nx
import plotly.graph_objects as go

# Define layers and example functionalities
layers = {
    "Presentation": {"description": "User interface, web pages, GUI"},
    "Application Logic": {"description": "Business logic, computations"},
    "Service": {"description": "Database, directory services, IPC"}
}

# Create directed graph
G = nx.DiGraph()
G.add_nodes_from(layers.keys())
G.add_edges_from([("Service", "Application Logic"), ("Application Logic", "Presentation")])

# Layout
pos = nx.spring_layout(G, seed=42)

# Edges
edge_x, edge_y = [], []
for u, v in G.edges():
    x0, y0 = pos[u]
    x1, y1 = pos[v]
    edge_x.extend([x0, x1, None])
    edge_y.extend([y0, y1, None])

edge_trace = go.Scatter(x=edge_x, y=edge_y, line=dict(width=2, color='#888'), mode='lines')

# Nodes
node_x, node_y, node_text = [], [], []
for node, data in G.nodes(data=True):
    x, y = pos[node]
    node_x.append(x)
    node_y.append(y)
    node_text.append(f"{node}: {layers[node]['description']}")

node_trace = go.Scatter(
    x=node_x, y=node_y, mode='markers+text',
    text=list(G.nodes), textposition="bottom center",
    hovertext=node_text, hoverinfo='text',
    marker=dict(size=60, color=['lightgreen','lightblue','lightpink'], line=dict(width=2, color='black'))
)

# Build figure
fig = go.Figure(data=[edge_trace, node_trace],
                layout=go.Layout(
                    title='Distributed Application Architecture',
                    showlegend=False,
                    hovermode='closest',
                    xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
                    yaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
                    margin=dict(b=20,l=5,r=5,t=40)
                ))
fig.show()
````

# Features of Distributed Application Visualization

- **Hover over nodes:** View the **layer description**
- **Dependency flow:** Shows **Service → Application Logic → Presentation**
- **Interactive visualization:** Suitable for **notebooks or web apps**
